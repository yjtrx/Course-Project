### 4)Write a function with name "Goal_diff_count" which should return their Goal Differences

###### Using the same function, find the team which has maximum and minimum goal difference. Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count():
    Goal_differences = df.GoalsFor - df.GoalsAgainst
    return Goal_differences
```

```python
df_diff = pd.DataFrame(df['Team'])
df_diff['Goal_diff'] = Goal_diff_count()
```

```python
df_diff.sort_values(by='Goal_diff',ascending=False)
```

```python
print("This is the team with the highest goal difference")
df_diff[df_diff.Goal_diff == df_diff.Goal_diff.max()]
```

```python
print("This is the team with the lowest goal difference")
df_diff[df_diff.Goal_diff == df_diff.Goal_diff.min()]
```
