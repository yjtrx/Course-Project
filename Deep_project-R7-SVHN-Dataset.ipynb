```
def pro_gress():
    import time
    from progressbar import AnimatedMarker,ProgressBar,Percentage,ETA,FileTransferSpeed,Bar
    widgets = ['Processing: ', Percentage(), ' ',
                  Bar(marker='=',left='|',right='|'),
                  ' ', ETA()]
    pbar = ProgressBar(widgets=widgets, maxval=500)
    for i in range(100,500+1,50):
        time.sleep(0.2)
        pbar.update(i)
    pbar.finish()
    print('Done!!')
print("Setting the version..")
show = pro_gress()    
%tensorflow_version 2.x
```

```
print("One-hot encoding the y_val labels")
show = pro_gress()
y_val1 = tf.keras.utils.to_categorical(y_val)
```

```
print("One-hot encoding the y_test labels")
show = pro_gress()
y_test1 = tf.keras.utils.to_categorical(y_test[()])
```

```
print("One-hot encoding the y_train labels")
show = pro_gress()
y_train1 = tf.keras.utils.to_categorical(y_train[()])
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
model = Sequential([layers.Flatten(input_shape=(32,32))])
model.add(layers.Dense(600,activation='relu',input_dim=1024))
model.add(layers.BatchNormalization())
model.add(layers.Dense(400,activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dense(100,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
print("Creating & compiling the model !!")
show = pro_gress()
```

```
model.summary()
```

```
model.fit(X_train1,y_train1,epochs = 40,validation_data=(X_test1,y_test1),batch_size=32)
```

```
y_pred = model.predict(X_test1)
show = pro_gress()
```

```
out = []
for val in y_pred:
  out.append(np.argmax(val))
print(out)  
```

```
from sklearn.metrics import accuracy_score,classification_report
import numpy as np
acc = accuracy_score(y_test,out)
report=classification_report(y_test,out)
print("This is the obtained accuracy {}%".format(acc*100))
print("This is the report with the detailed scoring of F1 scores,precision & accuracy:-")
print(report)
```
